The next step is to use the perceptron on simulated datasets. Each dataset is supposed to have 8 instances, 4 from each of two groups. My assignment isn't really graded, so I don't care about doing 4 per group and chose to use 15 per group to make it easier for me to trust it. I think the other students are supposed to show how they created their datasets, but I don't really care because mine isn't graded. I used numpy.random to generate two datasets x-values. I calculated y-values at y=4x-3, then calculated random numbers to add or subtract (depending on the group) to make them separate. I didn't realize they were supposed to be scaled -1 to 1, so I had to scale them and fit them to a -1 to 1 scale. That gave me two datasets (15 instances in each dataset) that were linearly seperable. I then swapped two values between sets to make a non-linearly seperable dataset. I saved these as csv files. 

```python
lin_sep_values = []
lin_sep_labels = []

with open('linearSeperableImaginaryData.csv', 'r') as file:
    reader = csv.reader(file)
    # Ignore the header lines
    next(reader)
    for line in reader:
        values = []
        for thing in line[:-1]:
            values.append(float(thing))
        lin_sep_values.append(values)
        lin_sep_labels.append(int(line[-1]))
        
print(lin_sep_labels)
lin_sep_values
```

```python
linSep = Perceptron(step_size=0.1) #max_iterations=10, shuffle=False, 
linSep.fit(lin_sep_values, lin_sep_labels)
weights = linSep.get_weights()
print(linSep.best_accuracy)
print(weights)
```

```python
linSep.best_epoch
```

To get the slope, you use -w1/w2, to get the intercept you use w3/w2. 

```python
slope = -(weights[0]/weights[1])
intercept = (-weights[2]/weights[1])
print(slope)
print(intercept)
```

```python
dataframe = pd.DataFrame(lin_sep_values).rename(columns={0:"X1", 1:"X2"})
dataframe['Label'] = lin_sep_labels
dataframe.head()
```

```python
label0 = dataframe[dataframe['Label']==0]
label1 = dataframe[dataframe['Label']==1]
```

```python
fig, tmp = plt.subplots()
ax = label0.plot.scatter(y = "X2", x = "X1", ax=tmp, color = 'g', label = "Group 0", s = 10)
ax.scatter(y = label1["X2"], x = label1["X1"], color = 'b', label="Group 1", s = 10)

#parameters = linear_model.params
x = np.linspace(-1, 1, 2)
y = intercept + slope*x
plt.plot(x,y,label="Division Line", linewidth=2, color='k')
plt.ylabel("X2", color='black')
plt.xlabel("X1", color='black')
plt.legend(loc='upper left')
```

```python
change_to_numeric_dict = []
for entry in dtypes:
    if entry == "REAL":
        change_to_numeric_dict.append('REAL')
    else:
        tmp_dict = {}
        i = 0
        for part in entry:
            tmp_dict[part] = i
            i += 1
        change_to_numeric_dict.append(tmp_dict)
        
change_to_numeric_dict
```

```python
numeric_data = []
for entry in voting_object['data']:
    individual_values = []
    for individual_response, individual_dict in zip(entry, change_to_numeric_dict):
        
        if individual_dict == 'REAL':
            individual_values.append(individual_response)
        else:
            individual_values.append(individual_dict[individual_response])
    numeric_data. append(individual_values) 
```
